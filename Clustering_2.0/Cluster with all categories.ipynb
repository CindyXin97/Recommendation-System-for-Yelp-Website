# 2. Data Prepared

Using Categories and pop_income datasets and merging them to a dataset called df.

```python
Categories = pd.read_csv('Category_per_Zip_Freq.csv')
```

```python
Income = pd.read_csv('Phoenix_Pop_Income.csv')
```

```python
Categories.head()
```

```python
IP = Income[['Households_Mean_income','Households_Median_income','ZIP_CODE','Total_Pop']]
IP = IP.set_index('ZIP_CODE')
IP = IP.rename(columns={'Total_Pop':'Population', 'Households_Mean_income':'Mean_Income', 'Households_Median_income':'Median_Income'})
```

```python
df = pd.merge(IP, Categories, left_index=True, right_on=Categories['postal_code'])
```

```python
df['postal_code'] = df['postal_code'].astype(str)
df = df.set_index('postal_code')
df = df.drop('key_0',axis=1)
```

```python
df.head()
```

Now we have the data set for clustering.

# 3. Feature Selection

We tried tSNE first, but it appears PCA did a better job. 

```python
scaler = preprocessing.StandardScaler().fit(df)
df_S=scaler.transform(df)
pca = PCA(66)
df_P=pca.fit_transform(df_S)
#Trans_X = TSNE(random_state=2498).fit_transform(data1)
eigenvalues = pca.explained_variance_ratio_
##plot explained variance over the number of compinents
n=66
plt.bar(np.arange(n), eigenvalues[:n].cumsum())
plt.xlabel("Number of components")
plt.ylabel("Explained Variance")
plt.show()
```

```python
for i in range(n):
    if list(eigenvalues[:i+1].cumsum())[i] >= 0.8:
        print('{} leading principal components are needed to cover 80%'.format(i+1))
        break
```

```python
df_C = df_P[:,:23]
```

We sue silhouette score to decide how many clusters should be there. It appears 2 clusters is an good option.

```python
range_n_clusters = [2, 3, 4, 5, 6, 7, 8, 9, 10]
for n_clusters in range_n_clusters: #try different numbers of clusters
    km = KMeans(n_clusters=n_clusters, random_state=2498)
    cluster_labels = km.fit_predict(df_C)
    #report average Silhouette score
    silhouette_avg = silhouette_score(df_C, cluster_labels)
    print("For n_clusters ={},".format(n_clusters)+" the average silhouette_score is :{}".format(silhouette_avg))
```

Visualize using the first two components

```python
plt.plot(np.array(df_C[:,0]), np.array(df_C[:,1]),'o', alpha=0.8, label=True)
```

# 4. Clustering: Using DBSCAN

We tried KMeans and GaussianMixture first, but it didn't do a good job. So we tried DBSCAN and tuned the eps to get a optimal result. It appears to be good. But since we don't need to decide cluster numbers, the silhouette score step could be seen as a simple validation.

```python
dbscan = DBSCAN(eps=32).fit_predict(df_C)
```

```python
palette = np.array(sns.color_palette("hls", 3))
f = plt.figure(figsize=(6, 6))
ax = plt.subplot(aspect='equal')
sc = ax.scatter(df_C[:,0], df_C[:,1], c=palette[dbscan])
ax.axis('tight')
```

Reassigning the labels back to df.

```python
df['label'] = dbscan
```

# 6. Folium Visualization

Now we use folium to build an interactive map.

Now we could plot it mapping all our data on.

```python
map_phoenix.choropleth(
    geo_data=gj,
    #name='Income Tax Per Capita',
    data=gdf,
    columns=['ZIP_CODE', 'labels'],
    fill_color= 'RdGy',
    key_on='properties.ZIP_CODE',
    fill_opacity=0.9,
    line_opacity=0.3,
    
) 
```

# 7. Adding Labels

We could add labels as popup in folium, which makes a it a better visualization.

```python
lat, lon = 0, 0
longitude = []
latitude = []
i = 0
while i < len(geometries):
    new = []
    temp = geometries[i][10:-2]
    temp = temp.split(', ')
    for x in temp:
        n = x.split(' ')
        new.append(n)
    j = 0
    while j < len(new):
        new[j][0] = float(new[j][0])
        new[j][1] = float(new[j][1])
        lon += new[j][0]
        lat += new[j][1]
        j = j + 1
    lon = lon/len(new)
    lat = lat/len(new)
    longitude.append(lon)
    latitude.append(lat)
    lon = lat = 0
    i = i + 1
```

Using HTML to set up the background

```python
table = """
<!DOCTYPE html>
<html>
<head>
<style>
table {{
    width:100%;
}}
table, th, td {{
    border: 1px solid black;
    border-collapse: collapse;
}}
th, td {{
    padding: 5px;
    text-align: left;
}}
table#t01 tr:nth-child(odd) {{
    background-color: #eee;
}}
table#t01 tr:nth-child(even) {{
   background-color:#fff;
}}
</style>
</head>
<body>

<table id="t01">
  <tr>
    <td>Zip Code</td>
    <td>{}</td>
  </tr>
  <tr>
    <td>Population</td>
    <td>{}</td>
  </tr>
  <tr>
    <td>Cluster Label</td>
    <td>{}</td>
  </tr>
</table>
</body>
</html>
""".format
```

```python
gdf = gdf.reset_index()
gdf = gdf.drop(['index'],axis=1)
```

```python
gdf['labels'] = gdf['labels'].astype(int)
```

Plot marks on the former layer.

```python
for i,c in gdf.iterrows():
    iframe = IFrame(html=table(c['ZIP_CODE'],c['POPULATION'],c['labels'])
                    ,width=310,height=110)
    popup = folium.Popup(iframe,max_width=400)
    
    folium.Marker(
        [latitude[i], longitude[i]],
        popup=popup,
        parse_html=True).add_to(map_phoenix) 
```
